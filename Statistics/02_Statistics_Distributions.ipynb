### PROBABILITY MASS FUNCTION

```python
# Import the required libraries
from scipy.stats import bernoulli
import matplotlib.pyplot as plt

# Instance of Bernoulli distribution with parameter p = 0.8
bd=bernoulli(0.2)
# Outcome of random variable either 0 or 1
x=[0,1]
y= bd.pmf(x)
# For the visualization of the bar plot of Bernoulli's distribution

plt.xlim(-1, 2)
plt.bar(x, y, color='blue')

# For labelling of Bar plot
plt.title('Bernoulli distribution (p=0.2)', fontsize='20')
plt.xlabel('Values of random variable x (0, 1)', fontsize='20')
plt.ylabel('Probability', fontsize='20')

plt.show()
```

### PROBABILITY DENSITY FUNCTION

```python
import numpy as np
from scipy.stats import norm

x = np.linspace(0.3, 1.75, 1000)
plt.plot(x, norm.pdf(x, 1.075, 0.2))
plt.show()
```

Probability density function's value at some specific point does not give you probability; it is a measure of how dense the distribution is around that value. For continuous random variables, the probability at a given point is equal to zero. Instead of p(X = x), we calculate probabilities between 2 points p(x1 < X < x2) and it is equal to the area below that probability density function. Probability density function's value can very well be above 1. It can even approach to infinity, but the are under the curve should be equal to 1

### CUMULATIVE DENSITY FUNCTION

```python
import numpy as np
from scipy.stats import norm

x = np.linspace(0.3, 1.75, 1000)
plt.plot(x, norm.cdf(x, 1.075, 0.2))
plt.show()
```

### PROBABILITY DENSITY FUNCTION

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as stats
mu, sigma = 3., 1. # mean and standard deviation
ln = np.random.lognormal(mu, sigma, 1000) # ln contains log normal data  (X axis)
p = stats.lognorm.pdf(x = ln, scale = mu, s = sigma) # Y axis
plt.plot(ln,p,'bo') # draw the pdf of lognormal
plt.show()
```

### CUMULATIVE DENSITY FUNCTION

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as stats
mu, sigma = 3., 1. # mean and standard deviation
ln = np.random.lognormal(mu, sigma, 1000) # ln contains log normal data  (X axis)
p = stats.lognorm.cdf(x = ln, scale = mu, s = sigma) # Y axis
plt.plot(ln,p,'ro') # draw the pdf of lognormal
plt.show()
```

EXAMPLE PROBLEM
        
        The average weight gained by a person over the winter months is uniformly distributed and ranges from 0 to 30 lbs. Find the probability of a person that he will gain between 10 and 15lbs in the winter months.
        

```python
from scipy.stats import uniform
import matplotlib.pyplot as plt
import numpy as np

#creating an array of values between
#-3 to 8 with a difference of 0.1
x = np.arange(0, 30, 1)
   
y = uniform.pdf(x, 5, 15)
plt.xlabel("weight in lbs")
plt.ylabel('probability of the weight')   
plt.plot(x, y) 
plt.show()
```

# BOX-COX TRANSFORMATION

        It transforms skwed data to normal distribution

```python
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
 
original_data = np.random.exponential(size = 1000) # generate non-normal data (exponential)
fitted_data, fitted_lambda = stats.boxcox(original_data) # transform training data & save lambda value
 
sns.distplot(original_data, hist = False) # POWER LAW DISTRIBUTION
sns.distplot(fitted_data, hist = False)  # NORMAL DISTRIBUTION
 
```
